```python
import pandas as pd 

goo = pd.read_csv("goo.csv",encoding='Big5') 
goo.head()
```

```python
# 篩選出 2002 年的資料
goo_2002 = goo[goo['年度'] == 2002]
```

```python
goo_sorted = goo_2002.sort_values(by=['英國', '法國', '德國', '西班牙', '瑞士', '荷蘭', '義大利', '奧地利', '比利時', '捷克', '瑞典', '丹麥', '芬蘭', '匈牙利', '愛爾蘭', '波蘭', '俄羅斯'], ascending=False)
# 选择前一项数据
top_country = goo_sorted.iloc[0:"country"]
```
