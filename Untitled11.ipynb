```python
import urllib.request as req
import json
import pandas as pd
url = "https://www.travel.com.tw/HALL/Index/EU"
request = req.Request(url, headers = {
    "user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36"
})
```

```python
with req.urlopen(request) as response:
    data = response.read().decode("utf-8")
#print (data)  #測試資料輸出
```

```python
import bs4
root = bs4.BeautifulSoup(data, "html.parser")
# print(root.title.string) 
```

```python
#尋找 ul標籤的 class_="release_list"
release_list = root.find_all("div", class_="col-sm-6")
print (release_list)  
```

```python
#轉存為Dataframe形式
df = pd.DataFrame(release_list)
print(df)
```

```python
#儲存為CSV檔
df.to_csv('travel_2023.csv')

print("csv檔成功下載")
```

```python
df.to_json("travel_2023.json") # 產生json檔
print("json檔成功下載")
```
