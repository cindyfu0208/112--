```python
#匯入json檔案，存入dict中
#此為可以拿到普發現金的地點
import json
player = open('goodd.json',encoding="utf8")
Data = json.load(player)
type(Data)
```

```python
temp = Data.get('atm')
temp
```

```python
ask = input("哪個區:")
for i in temp:
    if ask == i.get('鄉鎮縣市別'):
        print(ask, '有', i.get('裝設地點'),i.get('地址'),i.get('編號'))
```
